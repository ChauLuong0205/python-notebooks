Data Visualization with Pandas

This notebook explores simple data visualization techniques using pandas and matplotlib.  
Covered topics:

- Line, bar, pie, histogram plots using `.plot()`
- Customizing charts (colors, titles, labels, legend)
- Visualizing categorical vs numerical data
- Saving charts to file

 Goal: Communicate data trends effectively using visual tools.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\STUDY\Test_Python\scraping_web_example\Ice Cream Ratings.csv')
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('bmh')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings',
       ylabel = 'Scores')
```

```python
df.plot.barh( stacked = True)
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100 , c = 'Yellow')
```

```python
df.plot.hist(bins = 10)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10,8))
```
